```python
import os
import json
import argparse

hammer_path = "hammer_eval"

successful_lemma_names = []
failed_lemma_names = []

for file in os.listdir(hammer_path):
    if file.endswith("out"):
        result = open(os.path.join(hammer_path, file)).read().strip()
        info_path = os.path.join(hammer_path, file.replace("out", "info"))
        theorem_name = open(info_path).readlines()[-1].strip()
        if result == "true":
            successful_lemma_names.append(theorem_name)
        elif result == "false":
            failed_lemma_names.append(theorem_name)
        else:
            raise AssertionError
```

```python
total_proofs = 3000
eval_dir_names = [
    "paper_800m_ablation_vanilla"
]
```

```python
# Find the top 400 problem declarations
import json
proof_names = []
for i in range(total_proofs):
    file_path = f"universal_test_theorems/test_name_{i}.json"
    file_dict = json.load(open(file_path))
    proof_names.append(file_dict[0][1])
```

```python
proved_scripts = dict()

for eval_dir_name in eval_dir_names:
    new_script = []
    success = 0
    queue_empty = 0
    query_limit = 0
    step_timeout_limit = 0
    illegal_proof = 0
    timeout = 0
    total = 0

    for file in os.listdir("paper_800m_ablation_vanilla"):
        if file.endswith("json"):
            content = json.load(open(os.path.join("paper_800m_ablation_vanilla", file)))
            if "result" not in content["agent_info"]:
                continue
            total += 1
            if content["proved"] and "\noops" not in content["proof_script"] and "sorry" not in content["proof_script"]:
                success += 1
                new_script.append(content["proof_script"])
                print(content["proof_script"])
                print("#" * 150)
                continue
            if content["agent_info"]["result"] == "query limit reached":
                query_limit += 1
                continue
            if content["agent_info"]["result"] == "timeout":
                timeout += 1
                continue
            if content["agent_info"]["result"] == "queue empty":
                queue_empty += 1
                continue
            if content["agent_info"]["result"] == "step timeout limit reached":
                step_timeout_limit += 1
                continue
            if "\noops" in content["proof_script"] or "sorry" not in content["proof_script"]:
                illegal_proof += 1
    proved_scripts[eval_dir_name] = new_script

    assert total == success + query_limit + timeout + queue_empty + step_timeout_limit + illegal_proof
    print(f"Total: {total}, success: {success/total}, query limit reached: {query_limit/total}, timeout: {timeout/total}, "
          f"queue empty: {queue_empty/total}, step timeout limit: {step_timeout_limit/total}, illegal proof: {illegal_proof/total}.")
    os.system(delete_command)
```

```python
names_to_scripts = dict()
for key, value in proved_scripts.items():
    single_names_to_scripts = dict()
    for lemma in value:
        single_names_to_scripts[lemma.split("\n")[0].strip()] = lemma
        
    names_to_scripts[key] = single_names_to_scripts
```

```python
lm_success_lemma_names = [[lemma.split("\n")[0].strip() for lemma in lemmas] for lemmas in proved_scripts.values()][0]
```
